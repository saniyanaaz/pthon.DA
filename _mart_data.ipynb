```python
plt.figure(figsize=(10, 5))
df.groupby('CATEGORY')['SALES AMOUNT'].sum().sort_values().plot(kind='barh')
plt.title('Total Sales Amount by Product Category')
plt.xlabel('Sales Amount')
plt.ylabel('Product Category')
plt.grid(axis='x')
plt.show()
```

```python

plt.figure(figsize=(10, 5))
plt.plot(df['DATE'], df['SALES AMOUNT'], marker='o', linestyle='-')
plt.title('Sales Amount over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(df['PURCHASE PRICE'], df['SALE PRICE'], alpha=0.5)
plt.title('Scatter Plot: Purchase Price vs. Sale Price')
plt.xlabel('Purchase Price')
plt.ylabel('Sale Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='CATEGORY', y='SALES AMOUNT', data=df)
plt.title('Box Plot: Sales Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Sales Amount')
plt.show()
```

```python

plt.figure(figsize=(10, 5))
sns.histplot(df['SALES AMOUNT'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Sales Amount')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
x = df1["Date"]
y = df1['Close']

plt.figure(figsize=(10, 5))
sns.set(style="whitegrid")
step = len(x) // 10
plt.xticks(x[::step])
plt.plot(x, y, label="Closing Price", color="green")
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("AAPL Stock Price Data")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
